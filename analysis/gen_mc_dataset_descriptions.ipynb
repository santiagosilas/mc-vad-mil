```python
import os, sys
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
path_base = "/media/dev/Silas-PhD-Backup/MyPhD-ProcData/"

path_npy_pets_train = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/pets2009-relabeled-i3d/pets2009-combine-seed-1-train-view-002.npy")
path_npy_pets_test = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/pets2009-relabeled-i3d/pets2009-combine-seed-1-test-view-002.npy")

path_npy_upfl_train = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/upfall-i3d/upfall-combine-train-single-camera-c1-seed1.npy")
path_npy_upfl_test = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/upfall-i3d/upfall-combine-test-single-camera-c1-seed1.npy")

path_npy_hqfs_train = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/hqfs-relabeled-i3d/hqfs-i3d-rgb-train-cam2.npy")
path_npy_hqfs_test = os.path.join(f"/media/dev/LaCie/MC-VAD-MIL-DATA/center-crop/hqfs-relabeled-i3d/hqfs-i3d-rgb-test-cam2.npy")
```

```python
bags_upfl_train = np.load(path_npy_upfl_train, allow_pickle = True)
bags_upfl_test = np.load(path_npy_upfl_test, allow_pickle = True)
```

```python
pd.set_option('display.max_rows', None)
```

```python
table = list()
for item in bags_upfl_train:

    n = item["name"]
    n = n.replace("Subject","S")
    n = n.replace("Activity","-A")
    n = n.replace("Trial","-T")
    
    table.append([
        n, "N" if sum(item["y_fi"]) == 0. else "A", item["X_i"].shape[0], item["X_i"].shape[0] * 16
    ])
display(pd.DataFrame(table, columns=["Training Sequence","C","S","F"]))
```

```python
table = list()
for item in bags_upfl_test:

    n = item["name"]
    n = n.replace("Subject","S")
    n = n.replace("Activity","-A")
    n = n.replace("Trial","-T")
    
    table.append([
        n, "N" if sum(item["y_fi"]) == 0. else "A", item["X_i"].shape[0], item["X_i"].shape[0] * 16
    ])
display(pd.DataFrame(table, columns=["Test Sequence","C","S","F"]))
```
