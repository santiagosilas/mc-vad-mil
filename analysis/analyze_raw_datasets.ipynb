```python
# Check Datasets
import os
import cv2
import numpy as np
import pandas as pd

def count_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()
    return total_frames


# Folder with dataset subfolders
dir_datasets = "/media/dev/SILAS-SAND"
```

```python
# Filepaths for the Pets dataset
filepaths = list()
dir_dataset_name = "MP4-PETS"
dir_dataset = os.path.join(dir_datasets, dir_dataset_name)

for f in os.listdir(os.path.join(dir_dataset)):
    filepath = os.path.join(dir_datasets, dir_dataset, f)
    filepaths.append(filepath)
print(f"{dir_dataset_name}\t{len(filepaths)} videos.")

table_counts = list()
count = 0
for f in os.listdir(os.path.join(dir_dataset)):
  camera = int(f.split("-View_")[-1].replace(".mp4",""))
  scene = f.split("-View_")[0]
  filepath = os.path.join(dir_datasets, dir_dataset, f)
  num_frames_video = count_frames(filepath)
  table_counts.append([scene, camera, num_frames_video])
  print(count+1,end=",")
  count += 1

table_counts_pets = table_counts
```

```python
# Filepaths for the UpFall dataset
filepaths = list()
dir_dataset_name = "MP4-UPFALL"
dir_dataset = os.path.join(dir_datasets, dir_dataset_name)

for f in os.listdir(os.path.join(dir_dataset)):
    filepath = os.path.join(dir_datasets, dir_dataset, f)
    filepaths.append(filepath)
print(f"{dir_dataset_name}\t{len(filepaths)} videos.")

table_counts = list()
count = 0
for f in os.listdir(os.path.join(dir_dataset)):
  camera = int(f.split("Camera")[-1].replace(".mp4",""))
  scene = f.split("Camera")[0].replace(".mp4","")
  filepath = os.path.join(dir_datasets, dir_dataset, f)
  num_frames_video = count_frames(filepath)
  table_counts.append([scene, camera, num_frames_video])
  print(count+1,end=",")
  count += 1

table_counts_upfall = table_counts
```

```python
# Filepaths for the HQFS dataset
filepaths = list()
dir_dataset_name = "MP4-HQFS"
dir_dataset = os.path.join(dir_datasets, dir_dataset_name)

for f in os.listdir(os.path.join(dir_dataset)):
    filepath = os.path.join(dir_datasets, dir_dataset, f)
    filepaths.append(filepath)
print(f"{dir_dataset_name}\t{len(filepaths)} videos.")

table_counts = list()
count = 0
for f in os.listdir(os.path.join(dir_dataset)):
  camera = int(f.replace("cam","Cam").split("_Cam")[-1].replace(".avi",""))
  scene = f.replace("cam","Cam").split("_Cam")[0].replace(".avi","")
  if scene == "Fall34" or scene == "Fall40" or scene == "Fall46":
        continue
  filepath = os.path.join(dir_datasets, dir_dataset, f)
  num_frames_video = count_frames(filepath)
  table_counts.append([scene, camera, num_frames_video])
  print(count+1,end=",")
  count += 1

table_counts_hqfs = table_counts
```
